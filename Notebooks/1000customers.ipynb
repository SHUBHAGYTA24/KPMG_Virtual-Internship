```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1= pd.read_excel("KPMG_final.xlsx",sheet_name="CustomerDemographic",header=1)
```

```python
df2= pd.read_excel("KPMG_final.xlsx",sheet_name="NewCustomerList",header=1)
```

```python
df_new= pd.concat([df1,df2],axis=0)
```

```python
df_new1=df_new.drop(columns=['first_name','last_name','deceased_indicator','country','property_valuation','default','Unnamed: 16','Unnamed: 17','Unnamed: 18','Unnamed: 19','Unnamed: 20','Rank','Rank','Rank','Value','tenure','postcode','address'],axis=1)
```

```python
import datetime
df_new1['Age']=df_new1['DOB'].apply(lambda x: pd.to_datetime('today').year-x.year)
```

```python
df_new1
```

```python
df_new2=df_new1.drop(['DOB'],axis=1)
```

```python
df_new2.head()
```

```python
def AGE_GROUPER(age):
    if age<=25:
        return "young"
    elif age>25 and age<=50:
        return "adult"
    elif age>50 and age<=75:
        return "senior citizen"
    else:
        return "old"
    
```

```python
df_new2['AGE_GROUP']=df_new2['Age'].apply(lambda x: AGE_GROUPER(x))
df_new2
```

```python
df_new3=df_new2.drop(['Age'],axis=1)
```

```python
df_new3.head()
```

```python
def corr(sex):
    if sex[0]=="F":
        return "F"
    if sex[0]=="M":
        return "M"
    if sex[0]=="U":
        return "U"
```

```python
df_new3['gender']=df_new3['gender'].apply(lambda x:corr(x))
```
